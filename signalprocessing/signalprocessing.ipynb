- signal prcessing is to manipulate signal in time domain to frequency domain
- Signals existing in nature are continuous (analog) signals.

## Basics of signal Processing 

- __Sampling__ : reduction of a continuous-time signal in to a discrete time signal
- __Quantization__ : process of mapping input signal value into a given countable number of levels 
- __Encoding__ : After quantization, each levels are represented into a sequence of bits that can be decoded. if the quantization level is `L` , the number of bits is `log2(L)`

```python
# import packages 

import numpy as np 
import matplotlib.pyplot as plt 

import scipy
from scipy import signal 
```

```python
t = np.arange(0, 11) 
x = (0.85) ** t 
```

```python
plt.plot(x)
```

### Continous Signal

```python
plt.figure(figsize=(10, 8)) 

# 1. Plotting Analog signal 
plt.subplot(2, 2, 1) 
plt.title("Analog Signal", fontsize=20)

plt.plot(t, x, linewidth=3, label='x(t) = (0.85)^t')
plt.xlabel('t', fontsize=15)
plt.ylabel('amplitude', fontsize=15)
plt.legend(loc='upper right') 

# 2. Sampling and plotting of Sample signal 
plt.subplot(2, 2, 2) 
plt.title("Sampling", fontsize=20) 
plt.plot(t, x, linewidth=3, label='x(t) = (0.85)^t')
n = t 

markerline, stemlines, baseline = plt.stem(n, x, label='x(n)=(0.85)^n')
plt.setp(stemlines, 'linewidth', 3) 
plt.xlabel('n', fontsize=15)
plt.ylabel('amplitude', fontsize=15)
plt.legend(loc='upper right') 

# 3. Quantization 
plt.subplot(2, 2, 3) 
plt.title("Quantization", fontsize=20) 

markerline, stemlines, baseline = plt.stem(n, x)
plt.setp(stemlines, 'linewidth', 3)
plt.xlabel("n", fontsize=15)
plt.ylabel("Range of Quantizer", fontsize=15) 
[plt.axhline(y = (i+1) / 10, xmin = 0, xmax = 10, color = 'k', linewidth = 1, linestyle='--') for i in range(10)]

plt.subplot(2, 2, 4) 
plt.title("Quantized signal", fontsize=20) 
xq = np.around(x, 1)
markerline, stemlines, baseline = plt.stem(n, xq)
plt.xlabel("n", fontsize=15)
plt.ylabel("Rangge of Quantizer", fontsize=15)
[plt.axhline(y = (i+1) / 10, xmin = 0, xmax = 10, color = 'k', linewidth = 1, linestyle='--') for i in range(10)]
plt.tight_layout()
plt.show()
```
